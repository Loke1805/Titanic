# Visual Explorations (EDA)

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x=df['Age'])
plt.title("Age Outliers")
plt.show()
```

```python
plt.figure(figsize=(10,6))

numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Passenger Count by Sex")
plt.show()

sns.countplot(x='Pclass', data=df)
plt.title("Passenger Count by Class")
plt.show()
```

```python
sns.pairplot(df[['Age','Fare','Survived']])
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title("Age vs Fare")
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Pclass")
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', kde=True)
plt.title("Age vs Survived")
plt.show()
```

# Observations

```python
1. Numerical Insights
*Age distribution is right-skewed; most passengers are young adults.
*Fare has extreme outliers; some passengers paid very high amounts.

2. Categorical Insights
*Most passengers belonged to 3rd class.
*More males than females onboard.

3 .Survival Insights
*Female survival rate is much higher.    
*1st class passengers survived more.
*Children under 10 show higher survival rate.
```

# Final Summary of Findings
